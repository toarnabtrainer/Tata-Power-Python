## **Tkinter GUI support in Jupyter Notebook**

```python
# Enable Tkinter GUI support in Jupyter Notebook
%gui tk

# Import required tkinter modules
import tkinter as tk
from tkinter import messagebox

# If this cell is run again, close the previous Tkinter window
try:
    root.destroy()
except Exception:
    pass


# Function to perform calculation
def calculate_result():
    try:
        # Get the first number from the first text box
        num1 = float(entry_num1.get())

        # Get the second number from the second text box
        num2 = float(entry_num2.get())

        # Get the selected operation from the radio buttons
        operation = selected_operation.get()

        # Perform addition
        if operation == "Addition":
            result = num1 + num2
            result_var.set(f"Result: {num1} + {num2} = {result}")

        # Perform subtraction
        elif operation == "Subtraction":
            result = num1 - num2
            result_var.set(f"Result: {num1} - {num2} = {result}")

        # Perform multiplication
        elif operation == "Multiplication":
            result = num1 * num2
            result_var.set(f"Result: {num1} × {num2} = {result}")

        # Perform division
        elif operation == "Division":
            # Check division by zero error
            if num2 == 0:
                messagebox.showerror("Math Error", "Division by zero is not allowed.")
                result_var.set("Result: Division by zero error")
            else:
                result = num1 / num2
                result_var.set(f"Result: {num1} ÷ {num2} = {result}")

    except ValueError:
        # Show error if user enters invalid number
        messagebox.showerror("Input Error", "Please enter valid numeric values.")
        result_var.set("Result: Invalid input")


# Create the main application window
root = tk.Tk()

# Set the title of the window
root.title("Simple Calculator")

# Increase window size so that result is visible
root.geometry("450x450")

# Optional: prevent the window from becoming too small
root.minsize(450, 450)

# Create heading label
heading_label = tk.Label(
    root,
    text="Simple Calculator",
    font=("Arial", 16, "bold")
)
heading_label.pack(pady=10)

# Create label for first number
label_num1 = tk.Label(root, text="Enter First Number:")
label_num1.pack()

# Create text box for first number
entry_num1 = tk.Entry(root, width=30)
entry_num1.pack(pady=5)

# Create label for second number
label_num2 = tk.Label(root, text="Enter Second Number:")
label_num2.pack()

# Create text box for second number
entry_num2 = tk.Entry(root, width=30)
entry_num2.pack(pady=5)

# Create variable to store selected operation
selected_operation = tk.StringVar()

# Set default selected operation
selected_operation.set("Addition")

# Create label for operation selection
operation_label = tk.Label(root, text="Select Operation:")
operation_label.pack(pady=5)

# Create radio buttons for arithmetic operations
tk.Radiobutton(
    root,
    text="Addition",
    variable=selected_operation,
    value="Addition"
).pack()

tk.Radiobutton(
    root,
    text="Subtraction",
    variable=selected_operation,
    value="Subtraction"
).pack()

tk.Radiobutton(
    root,
    text="Multiplication",
    variable=selected_operation,
    value="Multiplication"
).pack()

tk.Radiobutton(
    root,
    text="Division",
    variable=selected_operation,
    value="Division"
).pack()

# Create Calculate button
calculate_button = tk.Button(
    root,
    text="Calculate",
    command=calculate_result,
    bg="lightblue",
    width=15
)
calculate_button.pack(pady=15)

# Create variable to store and display result
result_var = tk.StringVar()

# Default result message
result_var.set("Result will be displayed here")

# Create result label
result_label = tk.Label(
    root,
    textvariable=result_var,
    font=("Arial", 12, "bold"),
    fg="blue",
    wraplength=400
)
result_label.pack(pady=10)

# In Jupyter Notebook, do not write root.mainloop()
# The command %gui tk already handles the Tkinter event loop
```
