```python
class MyClass:
    """This is a Doc String..."""
    class_var1 = 100  # static variable
    class_var2 = 200
    def __init__(self, data1, data2):  # constructor method
        print("Executing the constructor method...")
        self.inst_var1 = data1  # instance variable
        self.inst_var2 = data2  # self is object binding variable
    def update(self):
        print("Updating static/class variables...")
        MyClass.class_var1 += 50
        MyClass.class_var2 += 50
    def display(self):  # class method
        print("Executing the display method...")
        print(f"self = {self}...")
        print(f"Instance variables inst_var1 = {self.inst_var1} and inst_var2 = {self.inst_var2}...")
        print(f"Class variables class_var1 = {MyClass.class_var1} and inst_var2 = {MyClass.class_var2}...")
        print(f"Class variables class_var1 = {self.class_var1} and inst_var2 = {self.class_var2}...")
    def __del__(self):  # destructor method
        print("Destructor method is executing...")
ob1 = MyClass(111, 222)
ob1.display()
ob1.update()
ob2 = MyClass(444, 555)
ob2.display()
print(ob1.__doc__)
print(MyClass.__doc__)
```

```python
# There are three different types of methods: instance, class, static
class MyClass:
    classVar = 111  # class/static variable
    # declaring instance method
    def instMethod(self):
        print("Executing instance method...", self)
        self.instVar = 200
        MyClass.classVar = 222
        print(f"So instVar = {self.instVar}, classVar = {MyClass.classVar} and classVar = {MyClass.classVar}...")
    # declaring class method
    @classmethod    # annotation or decorator
    def classMethod(cla):
        print("Executing class method...", cla)
        cla.classVar = 300
        print(f"So classVar = {MyClass.classVar} and classVar = {cla.classVar}...")
    # declaring static method
    @staticmethod    # annotation or decorator
    def staticMethod():
        print("Executing static method...")
        MyClass.classVar = 400
        print(f"So classVar = {MyClass.classVar}...")
ob1 = MyClass()
ob1.instMethod()
# MyClass.instMethod()
ob1.classMethod()
MyClass.classMethod()
ob1.staticMethod()
MyClass.staticMethod()
```

```python
# built-in Python method
class MyClass1:
    def __init__(self):
        print("Constructor method is executing...")
class MyClass2:
    classVar1 = 100
    def __init__(self):
        self.instVar1 = 111
    def function1(self):
        print(f"So instVar1 = {self.instVar1}...")
class MyClass3(MyClass2):  # MyClass3 is inheriting properties from MyClass2
    '''This is a Document String...'''
    def function2(self):
        print("function2() is executing...")
ob1 = MyClass3()
print(isinstance(ob1, MyClass1), isinstance(ob1, MyClass2), isinstance(ob1, MyClass3))
print(hasattr(ob1, 'classVar1'), hasattr(ob1, 'myvar'), hasattr(ob1, 'function1'), hasattr(ob1, 'function2'))
print(getattr(ob1, 'classVar1'), getattr(ob1, 'instVar1'), getattr(ob1, 'function1'))
setattr(ob1, 'instVar1', 222); print(ob1.instVar1)
setattr(ob1, 'instVar2', 333); print(ob1.instVar2) # creating instance variable (only) if it is not pre-defined
print(issubclass(MyClass1, MyClass2), issubclass(MyClass2, MyClass3), issubclass(MyClass3, MyClass2))
print(vars(ob1))  # returns a dictionary of attributes of the object
print(dir(ob1))   # returns a list of all the attributes of the object
```

### **Inheritance**

```python
# single inheritance
class Base:
    def __init__(self):
        print("Base: Constructor method is executing...")
    def displayB(self):
        print("Base: Display method is executing...")
class Derived(Base):
    def __init__(self):
        print("Derived: Constructor method is executing...")
        super().__init__()  # way - 1
        Base.__init__(self)  # way - 2
        super(Derived, self).__init__()  # way - 3
    def displayD(self):
        print("Derived: Display method is executing...")
ob1 = Derived()
Base.__init__(ob1)  # way - 4
super(Derived, ob1).__init__()  # way - 5
ob1.displayB()
ob1.displayD()
```

```python
# multilevel inheritance
class Base:
    def __init__(self):
        print("Base: Constructor method is executing...")
    def displayB(self):
        print("Base: Display method is executing...")
    def function(self):
        print("Base: Function method is executing...")
class Derived1(Base):
    def displayD1(self):
        print("Derived1: DisplayD1 method executing...")
    def function(self):
        print("Derived1: Function method is executing...")
class Derived2(Derived1):
    def displayD2(self):
        print("Derived2: DisplayD2 method is executing...")
    def function(self):
        print("Derived2: Function method is executing...")
        Base.function(self)
ob1 = Derived2()
ob1.displayB()
ob1.displayD1()
ob1.displayD2()
print("--------------------------------------------")
ob1.function()
super(Derived2, ob1).function()
Base.function(ob1)
```

```python
# hierarchical inheritance
class Base:
    def __init__(self):
        print("Base: Constructor method is executing...")
    def displayB(self):
        print("Base: Display method is executing...")
    def function(self):
        print("Base: Function method is executing...")
class Derived1(Base):
    def __init__(self):
        print("Derived1: Constructor method is executing...")
    def displayD1(self):
        print("Derived1: Display method is executing...")
    def function(self):
        print("Derived1: Function method is executing...")
class Derived2(Base):
    def __init__(self):
        print("Derived2: Constructor method is executing...")
    def displayD2(self):
        print("Derived2: Display method is executing...")
    def function(self):
        print("Derived2: Function method is executing...")
ob1 = Derived1()
ob1.displayB()
ob1.displayD1()
ob1.function()
super(Derived1, ob1).function()
print("------------------------------------------------------")
ob2 = Derived2()
ob2.displayB()
ob2.displayD2()
ob2.function()
super(Derived2, ob2).function()
```

```python
# multiple inheritance
class Base1:
    def __init__(self):
        print("Base1: constructor method is executing...")
    def displayB1(self):
        print("Base1: DisplayB1 method is executing...")
    def function(self):
        print("Base1: Function method is executing...")
class Base2:
    def __init__(self):
        print("Base2: constructor method is executing...")
    def displayB2(self):
        print("Base2: Display method is executing...")
    def function(self):
        print("Base2: Function method is executing...")
class Derived(Base2, Base1):  # MRO: Method Resolution Order
    def displayD(self):
        print("Derived: Display method is executing...")
ob1 = Derived()
ob1.displayB1()
ob1.displayB2()
ob1.displayD()
ob1.function()
```

```python
# hybrid inheritance
class Base:
    def __init__(self):
        print("Base: Constructor method is executing...")
    def displayB(self):
        print("Base: Display method is executing...")
class Derived1(Base):
    def displayD1(self):
        print("Derived1: Display method is executing...")
class Derived2(Base):
    def displayD2(self):
        print("Derived2: Display method is executing...")
class Derived3(Derived1, Derived2):
    def displayD3(self):
        print("Derived3: Display method is executing...")
ob1 = Derived3()
ob1.displayB()
ob1.displayD1()
ob1.displayD2()
ob1.displayD3()
```

### **Operator Overloading**

```python
num1 = 100
num2 = 40
print(f"So {num1} + {num2} = {num1 + num2}...")
print(f"So {num1} + {num2} = {num1.__add__(num2)}...")
print(f"So {num1} - {num2} = {num1 - num2}...")
print(f"So {num1} - {num2} = {num1.__sub__(num2)}...")
print(f"So {num1} * {num2} = {num1.__mul__(num2)}...")
print(f"So {num1} / {num2} = {num1.__truediv__(num2)}...")
print(f"So {num1} // {num2} = {num1.__floordiv__(num2)}...")
print(f"So {num1} > {num2} = {num1.__gt__(num2)}...")
print(f"So {num1} <= {num2} = {num1.__le__(num2)}...")
```

```python
class MyClass:
    def __init__(self, xx, yy):
        self.x = xx
        self.y = yy
    def __add__(self, ob):
        print("Performing Addition Operation...")
        temp = MyClass(0, 0)
        temp.x = self.x + ob.x
        temp.y = self.y + ob.y
        return temp
    def __sub__(self, ob):
        print("Performing Subtraction Operation...")
        temp = MyClass(0, 0)
        temp.x = self.x - ob.x
        temp.y = self.y - ob.y
        return temp
ob1 = MyClass(100, 200)
ob2 = MyClass(11, 22)
print(f"ob1.x = {ob1.x}, ob1.y = {ob1.y} and ob2.x = {ob2.x}, ob2.y = {ob2.y}...")
result = ob1.__add__(ob2)
print(result.x, result.y)
result = ob1 + ob2
print(result.x, result.y)
result = ob1.__sub__(ob2)
print(result.x, result.y)
result = ob1 - ob2
print(result.x, result.y)
```

## **Abstract Class and Interfaces**

```python
# abstract class
from abc import ABC, abstractmethod   # here ABC => Abstract Base Class
class AbsBaseClass(ABC):
    def __init__(self):
        print("Abstract class constructor...")
    @abstractmethod   # annotation or decorator
    def abstractMethod(self):
        pass          # pass is a statement place holder
    def concreteMethod(self):
        print("concreteMethod() is executing...")
class Derived(AbsBaseClass):
    def abstractMethod(self):    # defining the abstract method
        print("So the abstract method got redefined...")
ob1 = Derived()
ob1.abstractMethod()
ob1.concreteMethod()
```

```python
# abstract class
from abc import ABC, abstractmethod   # here ABC => Abstract Base Class
class AbsBaseClass(ABC):
    def __init__(self):
        print("Abstract class constructor...")
    @abstractmethod   # annotation or decorator
    def abstractMethod(self):
        print("Initial content...")
    def concreteMethod(self):
        print("concreteMethod() is executing...")
class Derived(AbsBaseClass):
    def abstractMethod(self):    # defining the abstract method
        print("So the abstract method got redefined...")
        super().abstractMethod()
ob1 = Derived()
ob1.abstractMethod()
ob1.concreteMethod()
```

```python
# dealing with interface
from abc import ABC, abstractmethod   # ABC stands for Abstract Base Class
class interface(ABC):
    @abstractmethod
    def abstractMethod1(self):
        pass
    @abstractmethod
    def abstractMethod2(self):
        pass
    @abstractmethod
    def abstractMethod3(self):
        pass
class Derived(interface):
    def abstractMethod1(self):
        print("abstractMethod1() is redefined...")
    def abstractMethod2(self):
        print("abstractMethod2() is redefined...")
    def abstractMethod3(self):
        print("abstractMethod3() is redefined...")
ob1 = Derived()
ob1.abstractMethod1()
ob1.abstractMethod2()
ob1.abstractMethod3()
```
